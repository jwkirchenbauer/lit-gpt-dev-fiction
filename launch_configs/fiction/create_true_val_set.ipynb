```python
from datasets import load_from_disk
```

```python
ORIG_DS_PATH = "/p/vast1/pretrain/from_frontier/datasets/fiction/dolma_v1_6-sample/train_shuffled_old"

ds = load_from_disk(ORIG_DS_PATH)
ds
```

```python
num_rows = len(ds)
train_indices = list(range(int(num_rows*0.95)))
val_indices = list(range(train_indices[-1]+1, num_rows))

print(num_rows)
print(len(train_indices), train_indices[:5], train_indices[-5:])
print(len(val_indices), val_indices[:5], val_indices[-5:])
```

```python
train_ds = ds.select(train_indices)
val_ds = ds.select(val_indices)
```
